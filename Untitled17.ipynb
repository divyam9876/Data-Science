```python
with open('dataset.txt','r') as file:
  data=[]
  X=[]
  Y=[]
  for line in file:
    line=line.strip().replace('(','').replace(')','')
    x,y=line.split(',')
    X.append(float(x))
    Y.append(float(y))
    data.append((float(x),float(y)))

n=len(data)

x_total=0
y_total=0
xy_total=0

xSquare_total=0
for i in data:
  x_total+=i[0]
  y_total+=i[1]
  xy_total+=i[0]*i[1]
  xSquare_total+=i[0]**2

def LinearRegression_method():
  m=((n*xy_total)-(x_total*y_total))/((n*xSquare_total)-x_total**2)
  b=(y_total-(m*x_total))/n
  print('Intercept(b): ',b)
  print('Slope(m): ',m)

LinearRegression_method()
```
